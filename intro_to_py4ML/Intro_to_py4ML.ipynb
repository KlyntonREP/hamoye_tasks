```python
import pandas as pd
```

```python
df = pd.read_csv("FoodBalanceSheets_E_Africa_NOFLAG.csv", encoding="latin-1");
df.head()
```

```python
# Getting the year with the highest sum of stock variation
stock_variation = df.groupby(by=['Element'])[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].sum()
stock_variation
```

```python
corr = df[['Element Code','Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].corr()
corr
```

```python
# Total sum of wine produced in 2015 and 2018
wine_sum = df.groupby(by=['Item'])[['Y2015', 'Y2018']].sum()
wine_sum
```

```python
# This block of code helps to get the area with the highest sum in 2017
area = df.groupby(by="Area")[["Y2017"]].sum()
area.sort_values(by='Y2017').tail(7)
```

```python
# Getting the mean and standard deviation accross the whole dataset for the year 2017
std = df['Y2017'].std()
print("Standard deviation", std)

mean = df['Y2017'].mean()
print("Mean: ", mean)
```

```python
# Total supply of protein to madagascar in 2015
total_supply = df.groupby(by=["Element"])[["Y2015", "Area"]].sum()
print(total_supply)
```

```python
# Getting the total number of sum of processing in 2017
sum_processing = df.groupby(by=('Element'))['Y2017'].sum()
sum_processing
```

```python
# This block of code helps to get the area with the 7th lowest sum
area = df.groupby(by="Area")[["Y2017"]].sum()
area.sort_values(by='Y2017').head(7)
```

```python
# gives the length of uniques countries
unique_countries_count = len(df['Area'].unique()) 
print("Unique Country Count: ",unique_countries_count)


# lists the unique country names
unique_countries_names = df['Area'].unique() 
print("Unique Country Names: ",unique_countries_names)
```

```python
# Total number of missing data in 2014
total_missing = df["Y2014"].isnull().sum()
print("Total missing count: ", total_missing)


# formular to get missing percentage [ (total_missing / total_cells) * 100 ]
total_missing_percentage = (1589/60943) * 100
print("Totla missing percentage: ", total_missing_percentage)
```
